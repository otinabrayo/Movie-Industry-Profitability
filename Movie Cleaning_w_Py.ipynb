# Importing necessary packages

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head()
```

## Data Cleaning 

```python
movies.isnull().sum()
```

```python
# Lets drop nulls in the columns since its movies no need of data replacement in it

movies = movies.dropna()
```

```python
movies.dtypes
```

```python
movies['budget'].head()
```

```python
# movies['budget'] = movies['budget'].replace('')

movies['budget'] = movies['budget'].astype(float)
```

```python
# Since its a float type we just divide by 100 to get rid of the last 2 trailing zeros 

movies['budget'] = (movies['budget'] / 100).astype(int)
```

```python
# Droping unused column

movies.drop(columns =['original_language', 'spoken_languages', 'status', 'tagline', 'original_title'], inplace=True)
```

```python
movies.to_csv('tmdb_5000_movies_cleaned.csv', index=False)
```
